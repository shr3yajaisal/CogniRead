```python
sample = narrativeqa['test'][0]

document = sample['document']['text']
question = sample['question']['text']
reference_answer = sample['answers'][0]
```

```python
def chunk_text(text, chunk_size=4000):
    chunks = []
    while len(text) > chunk_size:
        idx = text.rfind(' ', 0, chunk_size)
        if idx == -1:
            idx = chunk_size
        chunks.append(text[:idx])
        text = text[idx:].strip()
    if text:
        chunks.append(text)
    return chunks

chunks = chunk_text(document, chunk_size=4000)
print(f"Number of chunks: {len(chunks)}")
```

```python
import torch
from transformers import pipeline, AutoTokenizer

summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0 if torch.cuda.is_available() else -1)

from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
tokens = tokenizer(document, return_tensors="pt", truncation=True, max_length=1024)
decoded_text = tokenizer.decode(tokens["input_ids"][0], skip_special_tokens=True)

gist_summary = summarizer(decoded_text, max_length=150, min_length=50, do_sample=False)[0]['summary_text']

print("Gist Memory Summary:\n", gist_summary)
```

```python
!pip install -q faiss-cpu sentence-transformers

from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

embedder = SentenceTransformer('all-MiniLM-L6-v2')

chunk_embeddings = embedder.encode(chunks)

dimension = chunk_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(chunk_embeddings))
```

```python
def retrieve_relevant_chunks(query, top_k=3):
    query_embedding = embedder.encode([query])
    _, indices = index.search(np.array(query_embedding), top_k)
    return [chunks[i] for i in indices[0]]

retrieved_chunks = retrieve_relevant_chunks(question, top_k=3)

print("Retrieved Chunks:\n")
for i, chunk in enumerate(retrieved_chunks):
    print(f"[Chunk {i+1}]\n{chunk[:300]}...\n")
```

```python
context = gist_summary + "\n\n" + "\n\n".join(retrieved_chunks)
```

```python
from transformers import pipeline

qa_model = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

result = qa_model(question=question, context=context)

print("Predicted Answer:\n", result['answer'])
print("\nReference Answer:\n", reference_answer)
```

```python
from evaluate import load

qa_metric = load("squad")

predicted_answer = result['answer'].strip()
reference_answer = sample['answers'][0]['text'].strip()

metric_result = qa_metric.compute(predictions=[{"prediction_text": predicted_answer, "id": "0"}],
                                  references=[{"answers": {"answer_start": [0], "text": [reference_answer]}, "id": "0"}])

print("F1 Score:", metric_result['f1'])
print("Exact Match:", metric_result['exact_match'])
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

pred_emb = model.encode(predicted_answer, convert_to_tensor=True)
ref_emb = model.encode(reference_answer, convert_to_tensor=True)

cos_sim = util.pytorch_cos_sim(pred_emb, ref_emb).item()

print("Cosine Similarity Score:", round(cos_sim, 4))
```

```python
def compute_precision(predicted, reference):
    pred_tokens = predicted.lower().split()
    ref_tokens = reference.lower().split()

    overlap = sum([1 for token in pred_tokens if token in ref_tokens])

    if len(pred_tokens) == 0:
        return 0.0
    return overlap / len(pred_tokens)
```

```python
def compute_recall(predicted, reference):
    pred_tokens = predicted.lower().split()
    ref_tokens = reference.lower().split()

    if len(ref_tokens) == 0:
        return 0.0
    overlap = sum([1 for token in ref_tokens if token in pred_tokens])

    return overlap / len(ref_tokens)
```

```python
def compute_f1(precision, recall):
    if precision + recall == 0:
        return 0.0
    return 2 * (precision * recall) / (precision + recall)
```

```python
predicted_answer = result['answer']
reference_answer = sample['answers'][0]['text']

precision = compute_precision(predicted_answer, reference_answer)
recall = compute_recall(predicted_answer, reference_answer)
custom_f1 = compute_f1(precision, recall)

print(f" Precision: {precision:.4f}")
print(f" Recall: {recall:.4f}")
print(f"  Custom F1 Score: {custom_f1:.4f}")
```
